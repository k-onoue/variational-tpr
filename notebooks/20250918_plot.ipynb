```python
import os

def count_csv_lines(directory_path):
    """
    指定されたディレクトリ内のすべてのCSVファイルの行数を数えます。

    Args:
        directory_path (str): CSVファイルが格納されているディレクトリへのパス。
    """
    try:
        # ディレクトリ内のすべてのファイルとディレクトリのリストを取得します
        files = os.listdir(directory_path)
        
        # CSVファイルのみをフィルタリングします
        csv_files = [f for f in files if f.endswith('.csv')]
        
        if not csv_files:
            print(f"ディレクトリ '{directory_path}' にCSVファイルが見つかりません。")
            return
            
        print(f"'{directory_path}' 内のCSVファイルの行数:")
        # 各CSVファイルの行数を数えます
        for filename in sorted(csv_files):
            filepath = os.path.join(directory_path, filename)
            try:
                with open(filepath, 'r', encoding='utf-8') as f:
                    # ファイル内のすべての行を読み込み、その数を取得します
                    line_count = len(f.readlines())
                    print(f"- {filename}: {line_count} 行")
            except Exception as e:
                print(f"{filename} の処理中にエラーが発生しました: {e}")

    except FileNotFoundError:
        print(f"エラー: ディレクトリ '{directory_path}' が見つかりません。")
    except Exception as e:
        print(f"エラーが発生しました: {e}")

# あなたの環境に合わせてディレクトリパスを調整してください
target_directory = '../results/20250917_195858/raw'
count_csv_lines(target_directory)
```
