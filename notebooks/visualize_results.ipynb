```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def visualize_results(csv_path="../results.csv", output_path="../results.png"):
    """
    Reads experiment results from a CSV file, generates a bar plot to compare
    model performance across datasets, and saves the plot to a file.

    Args:
        csv_path (str): The path to the input CSV file.
        output_path (str): The path to save the output plot image.
    """
    # Read the results data
    try:
        results_df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f"Error: The file {csv_path} was not found.")
        return

    # Set up the plot style
    sns.set_theme(style="whitegrid")

    # Create a figure and axes for the plot
    plt.figure(figsize=(14, 8))

    # Generate the bar plot
    # This plot will show the mean RMSE for each model on each dataset
    ax = sns.barplot(
        data=results_df,
        x="dataset",
        y="rmse",
        hue="model",
        palette="viridis",
        ci="sd"  # Show standard deviation as error bars
    )

    # Set plot title and labels
    ax.set_title("Model Performance Comparison (RMSE) Across Datasets", fontsize=16)
    ax.set_xlabel("Dataset", fontsize=12)
    ax.set_ylabel("Average RMSE", fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha="right")

    # Adjust layout to prevent labels from being cut off
    plt.tight_layout()

    # Save the plot to a file
    plt.savefig(output_path)

    print(f"Visualization saved to {output_path}")
    # To display the plot in an interactive environment (like a Jupyter notebook),
    # you would uncomment the following line:
    # plt.show()

if __name__ == "__main__":
    visualize_results()
```
